```python
d={'Cricket':[1,3,np.nan,4,6,7,2,np.nan],
   'Baseball':[5,np.nan,np.nan,5,7,4,7,3],
   'Tennis':[1,2,3,4,5,6,7,8]}
```

```python
type(d)
```

```python
df=pd.DataFrame(d)
df
#NaN - missing values
```

```python
#handling the missing values
#(1)Drop the rows or columns containing the missing values  (dropna)
#(2)Fill the missing values with 0 or mean or mode or median  (fillna)
```

```python
#1. Drop the rows containing the missing values
```

```python
df.dropna(axis=0)
```

```python
#2. Drop the columns containing the missing values
```

```python
df.dropna(axis=1)
```

```python
df
```

```python
df.fillna(value=0)
```

```python
df
```

```python
df['Cricket']=df['Cricket'].fillna(value=df['Cricket'].median())
```

```python
df
```

```python
df['Baseball']=df['Baseball'].fillna(value=df['Baseball'].median())
df
```

```python
df=df.set_index([['Alpha','Bravo','Charlie','Delta','Echo','Foxtrot','Golf','Hotel']])
#set_index is the function used to change the index values
df
```
